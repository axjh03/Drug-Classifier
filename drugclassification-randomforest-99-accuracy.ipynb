```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
```

```python
def CheckNullPoints(DataFrame_Name, Column_name, values=False):
    if ((DataFrame_Name[Column_name].isna().sum() > 0)and (values==True)):
        print(f"{Column_name} has {DataFrame_Name[Column_name].isna().sum()} null points")
    elif ((DataFrame_Name[Column_name].isna().sum() > 0)and values==False):
        print(f"{Column_name} has null points")
    else:
        print(f"{Column_name} doesn't have any null points")
```

```python
def differentStringFinder(df, columnName, uniqueValuesList):
    for i in range(len(df.index)):
        if df[columnName][i] not in uniqueValuesList:
            uniqueValuesList.append(df[columnName][i])
    return uniqueValuesList
```

```python
UniqueDrugNames = []
differentStringFinder(df, "Drug", UniqueDrugNames)
UniqueDrugNames
```

```python
UniqueCholestrol_values = []
differentStringFinder(df, "Cholesterol", UniqueCholestrol_values)
UniqueCholestrol_values
```

```python
UniqueBP_values = []
differentStringFinder(df, "BP", UniqueBP_values)
UniqueBP_values
```

```python
df.Sex,_ = pd.factorize(df.Sex)
df.BP,_ = pd.factorize(df.BP)
df.Cholesterol,_ = pd.factorize(df.Cholesterol)
```

```python
featuresToTrain = list(df.keys())
featuresToTrain.remove("Drug")
featuresToTrain
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[featuresToTrain], df["Drug"], train_size=0.8, shuffle=True)
```

```python
RF_model = RandomForestRegressor()
```

```python
RF_model.fit(X_train, y_train)
```

```python
y_pred = RF_model.predict(X_test)
```
